### **Problem Statement**

Have you ever wondered when the best time of year to book a hotel room is? Or the optimal length of stay in order to get the best daily rate? What if you wanted to predict whether or not a hotel was likely to receive a disproportionately high number of special requests? This hotel booking dataset can help you explore those questions!

This data set contains booking information for a city hotel and a resort hotel, and includes information such as when the booking was made, length of stay, the number of adults, children, and/or babies, and the number of available parking spaces, among other things. All personally identifying information has been removed from the data. Explore and analyse the data to discover important factors that govern the bookings.

The problem statement of that project is that we have to book a hotel/resort so then we need to check which one is best, depends on some of the aspect like what is the seasonality and what are the top performing months? What if you wanted to predict whether or not a hotel was likely to receive a disproportionately high number of special requests? and what are the top countries in terms of booking and revenue? what the most frequent guest type?

#### **Exploratory Data Analysis(EDA)**


```python
def annot_percent(plot):
    # Iterate through each bar in the plot
    for p in plot.patches:
        # Calculate the percentage
        percentage = '{:.1f}%'.format(100 * p.get_height() / len(data))
        # Annotate the bar with the percentage value
        plot.annotate(percentage,
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha = 'center', va = 'center',
                      xytext = (0, 9),
                      textcoords = 'offset points')
```

```python
# Plotting monthly cancellations and customer type
# Group the data by customer type and arrival date month, then calculate the cancellation count
monthly_cancellations = data.groupby(['customer_type', 'arrival_date_month'])['is_canceled'].sum().reset_index()
# Plot the barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='arrival_date_month', y='is_canceled', hue='customer_type', data=monthly_cancellations)

# Set the title and labels
plt.title('Monthly Cancellations by Customer Type')
plt.xlabel('Month of Arrival')
plt.ylabel('Number of Cancellations')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# the busiest month for hotel booking
df_not_canceled = data[data['is_canceled'] == 0]

# Order of the months
new_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Ensure the 'arrival_date_month' column is in the correct order
df_not_canceled['arrival_date_month'] = pd.Categorical(df_not_canceled['arrival_date_month'], categories=new_order, ordered=True)

# Count the number of bookings for each month
monthly_counts = df_not_canceled['arrival_date_month'].value_counts().reindex(new_order)

# Convert counts to percentages
monthly_percentages = monthly_counts / monthly_counts.sum() * 100

# Plotting
plt.figure(figsize=(18, 6))
sns.lineplot(x=monthly_percentages.index, y=monthly_percentages.values, marker='o')
plt.title('Busiest Month for Hotel Bookings')
plt.xlabel('Month')
plt.ylabel('Percentage of Bookings')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

above line plot gives a conclusion that most of the busiest months are July and August
